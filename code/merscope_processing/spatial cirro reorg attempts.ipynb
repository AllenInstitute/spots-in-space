```python
#create unique pt label within donor id
upt = {}
unique_pt = 0
for i in sorted(h5ad.obs.donor_pseudotime.unique()):
    for j in h5ad.obs[h5ad.obs['donor_pseudotime']==i].filename.unique():
        if j == 'x':
            print(j, unique_pt)
        upt[j] = unique_pt
        unique_pt+=1


h5ad.obs['unique_pseudotime'] = [upt[x] for x in h5ad.obs['filename']]
h5ad.obs
```

```python
#organize h5ad.obsm['spatial_cirro'] by h5ad.obs['donor_pseudotime'] 
import numpy as np
counter = 0
ycounter = 0
ydist = 20000
for ii, gb in h5ad.obs.groupby("unique_pseudotime"): #ii = section name, gb = indexed chunk of dataframe
    spatial = h5ad.obsm['spatial'][h5ad.obs.unique_pseudotime==ii,:]
    cirro_y = -(spatial[:,1] - np.mean(spatial[:,1])) - (ycounter*ydist)
    cirro_x = (spatial[:,0] - min(spatial[:,0])) +(counter*30000)
    counter += 1
    if counter % 10 == 0:
        ycounter += 1
        counter = 0
    coordinates_cirro = np.column_stack((cirro_x, cirro_y))
    h5ad.obsm['spatial_cirro'][np.where(h5ad.obs.unique_pseudotime==ii)] = coordinates_cirro
h5ad.write_h5ad('/home/imaging_mfish/surveyNAS05/scratch/human/AD_MTG/resegmented_data_copies/antibody_tech_reps_121222.h5ad')
```
