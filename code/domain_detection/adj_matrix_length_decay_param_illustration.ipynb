```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_points(n_points, x_range, y_range):
    # centered around the origin
    x = np.random.uniform(low=-x_range, high=x_range, size=n_points)
    y = np.random.uniform(low=-y_range, high=y_range, size=n_points)

    #store in 2d np array
    points = np.zeros((n_points, 2))
    points[:,0] = x
    points[:,1] = y

    return points

def generate_color_array(xy_coordinates, decay_rate):
    # Calculate the Euclidean distances
    distances = np.linalg.norm(xy_coordinates, axis=1)
    
    # Apply the color formula
    color_array = np.exp(-(distances**2) / (decay_rate**2))
    
    return color_array
```

```python
N_PTS = 100
X_RANGE = 10
Y_RANGE = X_RANGE
LEN_DECAY = 10

xy_coordinates = generate_random_points(N_PTS, X_RANGE, Y_RANGE)
xy_coordinates = np.append(xy_coordinates, [[0,0]], axis=0)
```

```python
len_decay = [1, 3, 5, 10]
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
for i, decay_rate in enumerate(len_decay):
    ax = axes[i]
    color_array = generate_color_array(xy_coordinates, decay_rate)
    sc = ax.scatter(xy_coordinates[:, 0], xy_coordinates[:, 1], c=color_array, cmap='Reds')
    # ax.axis('off')
    ax.set_aspect('equal')
    # ax.set_xlabel('X')
    # ax.set_ylabel('Y')
    spine_color = 'whitesmoke'
    ax.spines['left'].set_color(spine_color)
    ax.spines['right'].set_color(spine_color) 
    ax.spines['top'].set_color(spine_color) 
    ax.spines['bottom'].set_color(spine_color) 
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(-X_RANGE, X_RANGE)
    ax.set_ylim(-Y_RANGE, Y_RANGE)

plt.show()
```
